```python
import pandas as pd
```

```python
from pymongo import MongoClient
```

```python
import numpy as np
```

#### Connecting with mongodb

```python
client = MongoClient("localhost", 27017)
```

```python
db = client["assignment_1_sample_data"]
```

```python
cursor = db.books.find().limit(100)
```

#### Set option in panda to show 100 rows for dataframe

```python
pd.set_option('display.max_rows', 100)
```

#### Converting data to dataframe with defined column

```python
data = pd.DataFrame(list(cursor), columns = ['_id', "title", "pageCount", "shortDescription", "authors"])
```

```python
data
```

```python
csvFileName = "books.csv"
```

```python
data.to_csv(csvFileName, index=False, na_rep='Unkown')
```

#### saved the data to csv file in the workspace

```python
csvRead = pd.read_csv(csvFileName);
```

#### Using print to print the csv data

```python
%%time
print(csvRead)
```

```python
%%time
pd.DataFrame(csvRead)
```
